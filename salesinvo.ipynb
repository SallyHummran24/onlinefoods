```python
def plots(colname,Top10=False):
  prop=df.groupby(colname, as_index=False)['TotalAmount'].sum()
  prop['propration']=(prop.TotalAmount / df.TotalAmount.sum())*100
  #هنا نطلع النسبة المئوية لكل مجموع بالنسبة للمجموع الكلي
  prop=prop.sort_values('propration',ascending=False)
  print('Describing of all total amount values',prop.TotalAmount.describe())
  if Top10==True :
     prop=prop.head(10)


  fig1=px.bar(data_frame=prop,x=colname,y='TotalAmount')
  fig2=px.pie(data_frame=prop,names=colname,values='TotalAmount')

  fig1.show()
  fig2.show()
```

```python
plots('Country')
```

```python
#create dictionary first name +last name // dic=key|+value
cust_df ={
    'full_name': df.FirstName +' ' + df.LastName ,
    'totalamount' : df.TotalAmount
}
cust_df=pd.DataFrame(cust_df)
cust_df.head()
```

```python
fig=px.bar(cust_df.groupby('full_name')['totalamount'].count().sort_values(ascending=False).head(10),
text_auto= True , title='number of orders per customer',
labels={'Full Name':'full_name','value':'frequency'})
fig.update_traces(marker_color='rgb(200,200,100)', marker_line_color='rgb(200,200,100)', marker_line_width=2.5,
                  opacity=0.6,textposition='inside'
                 )
```
