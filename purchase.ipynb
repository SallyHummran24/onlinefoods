```python
plt.figure(figsize=(10, 6))
sns.histplot(df['EstimatedSalary'], bins=20, kde=True, color='green')
plt.title(' Estimated Salary for purchases')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Purchased',y='Age',data=df)
plt.title('Purchased vs Age')
plt.show()
```

```python
sns.boxplot(x='Purchased',y='EstimatedSalary',data=df)
plt.title('Purchased v/s EstimatedSalary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=df, )
plt.title(' Age vs. Estimated Salary  by Purchased')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```
